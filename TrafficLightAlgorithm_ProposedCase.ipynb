```python
import random
import matplotlib.pyplot as plt
import numpy as np

waiting_vehicles_straight_lane = [10,7,8,6] 
waiting_vehicles_right_lane = [5,3,4,2]
passed_vehicles_straight_lane = [0,0,0,0]
passed_vehicles_right_lane = [0,0,0,0]

trafficLow = [0.2,0.3]
traffic = [0.4,0.5]
trafficHeavy = [0.6,0.7]

def Tgs(vehicles1, vehicles2):
  tgs = ((30/10)**0.5)* max(vehicles1,vehicles2)/2 + 3
  return tgs

def Tgr(vehicles):
  tgr = ((30/10)**0.5)*vehicles + 2
  return tgr

def addTraffic(trafficType1, trafficType2, trafficType3, trafficType4):
  waiting_vehicles_straight_lane[0] = waiting_vehicles_straight_lane[0] + random.choice(trafficType1)
  waiting_vehicles_straight_lane[1] = waiting_vehicles_straight_lane[1] + random.choice(trafficType2)
  waiting_vehicles_straight_lane[2] = waiting_vehicles_straight_lane[2] + random.choice(trafficType3)
  waiting_vehicles_straight_lane[3] = waiting_vehicles_straight_lane[3] + random.choice(trafficType4)

  waiting_vehicles_right_lane[0] = waiting_vehicles_right_lane[0] + random.choice(trafficType1)/2
  waiting_vehicles_right_lane[1] = waiting_vehicles_right_lane[1] + random.choice(trafficType2)/2
  waiting_vehicles_right_lane[2] = waiting_vehicles_right_lane[2] + random.choice(trafficType3)/2
  waiting_vehicles_right_lane[3] = waiting_vehicles_right_lane[3] + random.choice(trafficType4)/2  

  
def Ty(speedLimit):
  ty = (speedLimit/4)*(5/18)
  return ty

def calculatedCaseTime(caseNumber):
  case_Time = Tgs(waiting_vehicles_straight_lane[caseNumber-1], waiting_vehicles_straight_lane[caseNumber+1])
  if case_Time > 30:
    case_Time = 30
  return case_Time

def calculatedCaseTimeRight(caseNumber):
  case_Time = Tgr(waiting_vehicles_straight_lane[caseNumber-1])
  if case_Time > 25:
    case_Time = 25
  return case_Time


def printCase(case_number):
    print("total time Passed: ",round(t))
    print("vehicles passsed: ",round(passed_vehicles_straight_lane[case_number-1]))
    print("vehicles waiting lane1: ",round(waiting_vehicles_straight_lane[0]))
    print("vehicles waiting lane2: ",round(waiting_vehicles_straight_lane[1]))
    print("vehicles waiting lane3: ",round(waiting_vehicles_straight_lane[2]))
    print("vehicles waiting lane4: ",round(waiting_vehicles_straight_lane[3]))


    print("vehicles waiting right lane1: ",round(waiting_vehicles_right_lane[0]))    
    print("vehicles waiting right lane2: ",round(waiting_vehicles_right_lane[1]))
    print("vehicles waiting right lane3: ",round(waiting_vehicles_right_lane[2]))
    print("vehicles waiting right lane4: ",round(waiting_vehicles_right_lane[3]))
    print('\n')   

def printCaseAfterYellow():
    print("time now after yellow: ", round(t))
    print("vehicles waiting lane1: ",round(waiting_vehicles_straight_lane[0]))
    print("vehicles waiting lane2: ",round(waiting_vehicles_straight_lane[1]))
    print("vehicles waiting lane3: ",round(waiting_vehicles_straight_lane[2]))
    print("vehicles waiting lane4: ",round(waiting_vehicles_straight_lane[3]))


    print("vehicles waiting right lane1: ",round(waiting_vehicles_right_lane[0]))    
    print("vehicles waiting right lane2: ",round(waiting_vehicles_right_lane[1]))
    print("vehicles waiting right lane3: ",round(waiting_vehicles_right_lane[2]))
    print("vehicles waiting right lane4: ",round(waiting_vehicles_right_lane[3]))
    print('\n')

def yelloLight(t,i):
  
  if i == 'straight':
    t=t+Ty(60)
    addTraffic(traffic, traffic, traffic, traffic)
    printCaseAfterYellow()
    print('\n')
    waitLineStraight_1.append(round(waiting_vehicles_straight_lane[0]))
    waitLineStraight_2.append(round(waiting_vehicles_straight_lane[1]))
    waitLineStraight_3.append(round(waiting_vehicles_straight_lane[2]))
    waitLineStraight_4.append(round(waiting_vehicles_straight_lane[3]))
    print(waitLineStraight_1)
    print(waitLineStraight_2)
    print(waitLineStraight_3)
    print(waitLineStraight_4)
    print('\n')
  elif i== 'right':
    t=t+Ty(60)
    addTraffic(traffic, traffic, traffic, traffic)
    printCaseAfterYellow()
    print('\n')
    waitLineRight_1.append(round(waiting_vehicles_right_lane[0]))
    waitLineRight_2.append(round(waiting_vehicles_right_lane[1]))
    waitLineRight_3.append(round(waiting_vehicles_right_lane[2]))
    waitLineRight_4.append(round(waiting_vehicles_right_lane[3]))
    print(waitLineRight_1)
    print(waitLineRight_2)
    print(waitLineRight_3)
    print(waitLineRight_4)
    print('\n')


cases = [1,2]
t=0
counter = 0
run_no = 1
runTime=[]


waitLineStraight_1=[]
waitLineStraight_2=[]
waitLineStraight_3=[]
waitLineStraight_4=[]

waitLineRight_1=[]
waitLineRight_2=[]
waitLineRight_3=[]
waitLineRight_4=[]


while t<600:
  print("This is Run Number ",run_no)

  for i in cases:
    print("Case Number ", i)
    caseTime = calculatedCaseTime(i)
    wt = waiting_vehicles_straight_lane[i-1]

    while counter < caseTime:  
      if waiting_vehicles_straight_lane[i-1] < 0:
        waiting_vehicles_straight_lane[i-1] = 0
      else:
        waiting_vehicles_straight_lane[i-1] = waiting_vehicles_straight_lane[i-1] - 2

      if passed_vehicles_straight_lane[i-1] > wt:
        passed_vehicles_straight_lane[i-1] = wt
      else:
        passed_vehicles_straight_lane[i-1] = passed_vehicles_straight_lane[i-1] + 2

      if waiting_vehicles_straight_lane[i-1] < 0:
        waiting_vehicles_straight_lane[i-1] = 0


      if waiting_vehicles_straight_lane[i+1] < 0:
        waiting_vehicles_straight_lane[i+1] = 0
      else:
        waiting_vehicles_straight_lane[i+1] = waiting_vehicles_straight_lane[i+1] - 2

      if passed_vehicles_straight_lane[i+1] > wt:
        passed_vehicles_straight_lane[i+1] = wt
      else:
        passed_vehicles_straight_lane[i+1] = passed_vehicles_straight_lane[i+1] + 2

      if waiting_vehicles_straight_lane[i+1] < 0:
        waiting_vehicles_straight_lane[i+1] = 0

      
      addTraffic(traffic, traffic, traffic, traffic)
      t = t + 1
      counter = counter + 1

    print("time given to case ", i, ": ", round(caseTime))
    printCase(i)

    counter = 0

    yelloLight(t,'straight')

    runTime.append(round(t))
    print(runTime)
    print('\n')

    print("Case Number ", i+1)
    caseTime = calculatedCaseTimeRight(i+1)

    wt = waiting_vehicles_right_lane[2-i]

    while counter < caseTime:  
      if waiting_vehicles_right_lane[2-i] < 0:
        waiting_vehicles_right_lane[2-i] = 0
      else:
        waiting_vehicles_right_lane[2-i] = waiting_vehicles_right_lane[2-i] - 1

      if passed_vehicles_right_lane[2-i] > wt:
        passed_vehicles_right_lane[2-i] = wt
      else:
        passed_vehicles_right_lane[2-i] = passed_vehicles_right_lane[2-i] + 1

      if waiting_vehicles_right_lane[2-i] < 0:
        waiting_vehicles_right_lane[2-i] = 0

         
      if waiting_vehicles_right_lane[4-i] < 0:
        waiting_vehicles_right_lane[4-i] = 0
      else:
        waiting_vehicles_right_lane[4-i] = waiting_vehicles_right_lane[4-i] - 1

      if passed_vehicles_right_lane[4-i] > wt:
        passed_vehicles_right_lane[4-i] = wt
      else:
        passed_vehicles_right_lane[4-i] = passed_vehicles_right_lane[4-i] + 1

      if waiting_vehicles_right_lane[4-i] < 0:
        waiting_vehicles_right_lane[4-i] = 0

      addTraffic(traffic, traffic, traffic, traffic)
      t = t + 1
      counter = counter + 1

    print("time given to case ", i+1, ": ", round(caseTime))
    printCase(i)

    counter = 0

    yelloLight(t,'right')

    runTime.append(round(t))
    print(runTime)
    print('\n')


  run_no = run_no+1
  print('\n')
```

```python
def average(Lst):
  return sum(Lst) / len(Lst)

totalWait1 = waitLineStraight_1 + waitLineStraight_3 
totalWait2 = waitLineRight_1 + waitLineRight_3
totalWait3 = waitLineStraight_2 + waitLineStraight_4 
totalWait4 = waitLineRight_2 + waitLineRight_4

print('\n')
print('The average vehicles waiting at any second are: ',int(average((totalWait1))+ int(average(totalWait2))+ int(average(totalWait3))+ int(average(totalWait4)))/4)
print('\n')

plt.plot(runTime, totalWait1, linewidth = 2)
plt.plot(runTime, totalWait2, linewidth = 2)
plt.plot(runTime, totalWait3, linewidth = 2)
plt.plot(runTime, totalWait4, linewidth = 2)

```
